```python
# ==========================================================
# INFERENTIAL STATISTICS ANALYSIS IN DAV
# Google Colab Ready
# ==========================================================

# Step 1: Install Required Libraries
!pip install pandas numpy matplotlib seaborn scipy -q

# ==========================================================
# Step 2: Import Libraries
# ==========================================================

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

# ==========================================================
# Step 3: Create Dataset
# ==========================================================

data = {
    'Maths': [78, 85, 90, 76, 88, 95, 89, 84, 91, 87],
    'Science': [80, 86, 92, 75, 85, 96, 90, 83, 93, 88]
}

df = pd.DataFrame(data)

print("STUDENT MARKS DATASET")
print(df)

# ==========================================================
# Step 4: Descriptive Statistics
# ==========================================================

print("\n==============================")
print("DESCRIPTIVE STATISTICS")
print("==============================")

print("\nMean Values:")
print(df.mean())

print("\nStandard Deviation:")
print(df.std())

print("\nStatistical Summary:")
print(df.describe())

# ==========================================================
# Step 5: Inferential Statistics - T-Test
# ==========================================================

print("\n==============================")
print("HYPOTHESIS TESTING (T-TEST)")
print("==============================")

# Null Hypothesis (H0):
# There is no significant difference between Maths and Science marks.

# Alternative Hypothesis (H1):
# There is a significant difference between Maths and Science marks.

t_stat, p_value = stats.ttest_ind(
    df['Maths'],
    df['Science']
)

print("T-Statistic =", round(t_stat, 4))
print("P-Value =", round(p_value, 4))

alpha = 0.05

if p_value < alpha:
    print("Result: Reject Null Hypothesis")
    print("There is a significant difference between the groups.")
else:
    print("Result: Fail to Reject Null Hypothesis")
    print("There is NO significant difference between the groups.")

# ==========================================================
# Step 6: Correlation Analysis
# ==========================================================

print("\n==============================")
print("CORRELATION ANALYSIS")
print("==============================")

correlation = df['Maths'].corr(df['Science'])

print("Correlation Coefficient =", round(correlation, 4))

if correlation > 0:
    print("Positive Correlation")
elif correlation < 0:
    print("Negative Correlation")
else:
    print("No Correlation")

# ==========================================================
# Step 7: Scatter Plot Visualization
# ==========================================================

plt.figure(figsize=(7,5))

plt.scatter(
    df['Maths'],
    df['Science'],
    s=100
)

plt.title("Maths vs Science Marks")
plt.xlabel("Maths Marks")
plt.ylabel("Science Marks")
plt.grid(True)

plt.show()

# ==========================================================
# Step 8: Histogram Visualization
# ==========================================================

plt.figure(figsize=(7,5))

plt.hist(
    df['Maths'],
    bins=5
)

plt.title("Maths Marks Distribution")
plt.xlabel("Marks")
plt.ylabel("Frequency")

plt.show()

# ==========================================================
# Step 9: Box Plot Visualization
# ==========================================================

plt.figure(figsize=(7,5))

sns.boxplot(data=df)

plt.title("Box Plot of Marks")

plt.show()

# ==========================================================
# Step 10: Heatmap Visualization
# ==========================================================

plt.figure(figsize=(6,4))

sns.heatmap(
    df.corr(),
    annot=True,
    cmap='YlGnBu'
)

plt.title("Correlation Heatmap")

plt.show()

# ==========================================================
# Step 11: Pair Plot Visualization
# ==========================================================

sns.pairplot(df)
plt.show()

# ==========================================================
# Step 12: Final Output
# ==========================================================

print("\n========================================")
print("Inferential Statistics Analysis Completed Successfully!")
print("========================================")
```
